# How To Drow Sebaorn Scatter plot?

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset("titanic")
df
```

```python
sns.scatterplot(x="age",y="fare",data=df)
```

```python
sns.scatterplot(x="age",y="fare",data=df,hue="sex")
```

```python
sns.scatterplot(x="age",y="fare",data=df,hue="sex",style="who")
```

```python
sns.scatterplot(x="age",y="fare",data=df,hue="sex",style="who" ,size="who")
```

```python

sns.scatterplot(x="age",y="fare",data=df,hue="sex",style="who" ,size="who",sizes=(100,400))
```

```python
plt.figure(figsize=(16,9))
sns.scatterplot(x="age",y="fare",data=df,hue="sex",style="who" ,size="who",sizes=(100,400))
```

```python
plt.figure(figsize=(16,9))
sns.scatterplot(x="who",y="fare",data=df,hue="alive",style="alive" ,size="who",sizes=(100,400))
```

```python
plt.figure(figsize=(16,9))
sns.scatterplot(x="who",y="fare",data=df,hue="alive",style="alive" ,size="who",sizes=(100,400),
               palette="inferno_r",alpha=.7)
```

```python
plt.figure(figsize=(12,4)),
sns.scatterplot(x="age",y="fare",data=df)
sns.lineplot(x="age",y="fare",data=df)
sns.barplot(x="age",y="fare",data=df)
```

# How to draw Seaborn Heatmap?

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
 arr_2d = np.linspace(1,5,12).reshape(4,3)
arr_2d
  
```

```python
sns.heatmap(arr_2d)
```

```python
globalWarming_df = pd.read_csv('Who_is_responsible_for_global_warming.csv')
globalWarming_df.head()
```

```python
globalWarming_df=globalWarming_df.drop(columns=["Country Code","Indicator Name","Indicator Code"],axis=1).set_index("Country Name")
globalWarming_df.head(6)
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(globalWarming_df)
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(globalWarming_df,vmin=0,vmax=21)
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(globalWarming_df,cmap='coolwarm')
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(globalWarming_df,cmap='coolwarm',annot=True)
```

```python
annot_arr=np.array([['a00','a01','a02'],['a10','a11','a12'],['a21','a22','a23'],['a30','a31','a33']])
annot_arr
```

```python
sns.heatmap(arr_2d,annot=annot_arr,fmt="s")
```

```python
globalWarming_df.corr()
```

```python
plt.figure(figsize=(16,9))
ax=sns.heatmap(globalWarming_df.corr(),annot=True,linewidth=3)
ax.tick_params(size=10,color='w',labelsize=10,labelcolor='w')
plt.title("heatmap of 'who is responsible for global warming'",fontsize=25)
plt.show()
```
